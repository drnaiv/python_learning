```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Data Sciences\Project\Uncleaned_DS_jobs.csv\Uncleaned_DS_jobs.csv", header = 0, )
df
```

```python
df.drop(['index'], axis= 1)
```

```python
#removing unnecessary information from values in the Company Name, Job Description column
df['Job Description'] = df['Job Description'].str.replace('\n', ' ', regex = True)
df['Company Name'] = df['Company Name'].str.split('\n').str[0]
df
```

```python
df.info()
```

```python
df.drop(['Headquarters','Sector','Competitors'],axis=1)
df
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('K', '000')
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(Glassdoor est.)', '')
df['Salary Estimate'] = df['Salary Estimate'].str.replace('$', '')
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(Employer est.)', '')
df['Salary Estimate']
```

```python
df['Size'].unique()
```

```python
df['Size'] = df['Size'].str.replace('-1','Unknown')
```

```python
df['Founded'].unique()
```

```python
df['Founded'] = df['Founded'].replace(-1,0)
```

```python
df['Type of ownership'].unique()
```

```python
df['Type of ownership'] = df['Type of ownership'].str.replace('-1', 'Unknown')
```

```python
df['Industry'] = df['Industry'].str.replace('-1', 'Unknown')
```

```python
df['Revenue'].value_counts()
```

```python
df['Revenue'] = df['Revenue'].str.replace('-1', 'Unknown / Non-Applicable')
```

```python
df['Revenue'].value_counts()
```

```python
df.isnull().sum()
```

```python
df.drop(['index'], axis=1)
```
