```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# import library untuk ignore future warning
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
path_data = "D:\Data Sciences\Data Visualization & Wrangling\Project\AirBNB\Bangkok\listings.csv.gz"

listings = pd.read_csv(path_data)

listings.head()
```

```python
# Set pandas max columns jumlah kolom dari data
pd.set_option('display.max_columns', len(listings.columns))
pd.set_option('display.max_rows', 110)

# Menampilkan 5 data teratas
listings.head()
```

```python
# dimensi data set
row, columns = listings.shape

print(f"Listing terdiri dari {row} baris dan {columns} kolom")
```

```python
listings.host_since = pd.to_datetime(listings.host_since, yearfirst=True)

listings.host_since
```

```python
listings.host_response_rate.head()
```

```python
# Menghapus tanda percent
listings.host_response_rate = listings.host_response_rate.str.replace("%", "")
listings.host_response_rate
```

```python
# Mengisi nan dengan -1
listings.host_response_rate.fillna(-1, inplace=True)

# Mengubah tipe data menjadi int
listings.host_response_rate = listings.host_response_rate.astype(np.int64)

# mengembalikan -1 manjadi nan
listings.host_response_rate.replace(-1, np.nan, inplace=True)

# menampilkan data host_response_rate
listings.host_response_rate
```

```python
# Mengubah nama kolom sehingga represntatif
listings = listings.rename(columns = {'host_response_rate':'percent_host_response_rate'})
listings.percent_host_response_rate
```

```python
listings.host_acceptance_rate.head()
```

```python
# Menghapus tanda percent
listings.host_acceptance_rate = listings.host_acceptance_rate.str.replace("%", "")
listings.host_acceptance_rate
```

```python
# Mengisi nan dengan -1
listings.host_acceptance_rate.fillna(-1, inplace=True)

# Mengubah tipe data menjadi int
listings.host_acceptance_rate = listings.host_acceptance_rate.astype(np.int64)

# mengembalikan -1 manjadi nan
listings.host_acceptance_rate.replace(-1, np.nan, inplace=True)

# menampilkan data host_response_rate
listings.host_acceptance_rate
```

```python
# cek data pada kolom price
listings.price
```

```python
# menghilangkan tanda $ dan koma
listings.price = listings.price.str.replace("$", "")
listings.price = listings.price.str.replace(",", "")
listings.price = listings.price.str.replace(".00", "")


# mengubah kolom sehingga bertipe float
listings.price = listings.price.astype("float")

# Menampilkan data price
listings.price
```

```python
# Mendapatkan jumlah missing value tiap kolom
# dan mengurutkan dari yang terbesar ke terkecil
nan_col = listings.isna().sum().sort_values(ascending = False)
nan_col
```

```python
# Mendapatkan persentase missing value tiap kolom
n_data = len(listings)

percent_nan_col = (nan_col/n_data) * 100
percent_nan_col
```

```python
# Buat dalam bentuk dataframe
column_contain_nan = pd.DataFrame(percent_nan_col[percent_nan_col > 0],
                                  columns = ["%nan"])

column_contain_nan
```

```python
# Cari kolom yang persentase missing valuenya lebih dari 90%
nan_col_abv90 = percent_nan_col[percent_nan_col > 90].index
nan_col_abv90
```

```python
# Drop kolom yang %Nan > 90%
listings_filtered = listings.drop(columns = nan_col_abv90)

listings_filtered.head()
```

```python
# cek jumlah data unik pada property type
n_unique = listings.property_type.nunique()

print(f"Jumlah data unik pada kolom property type: {n_unique}")
```

```python
# cek data unik pada property type
listings.property_type.unique()
```

```python
# menyeragamkan kategori yang mengandung private room
listings.property_type = listings.property_type.apply(lambda x: 'private room' if 'Private room' in x else x)

# menyeragamkan kategori yang mengandung shared room
listings.property_type = listings.property_type.apply(lambda x: 'shared room' if 'Shared room' in x else x)

# menyeragamkan kategori yang mengandung Entire
listings.property_type = listings.property_type.apply(lambda x: 'entire home/apt' if 'Entire' in x else x)

# menyeragamkan kategori yang tidak mengandung Entire shared dan private menjadi others
listings.property_type = listings.property_type.apply(lambda x: 'others' if ('entire' not in x
                                                                                               and 'private' not in x
                                                                                               and 'shared' not in x) else x)
```

```python
# cek data unik pada property type
listings.property_type.unique()
```

```python
# data unik untuk kolom host_response_time
data_unik_host_response = listings["host_response_time"].unique()
print(f"data unik pada kolom host_response_time: {data_unik_host_response}")

# membuat spasi
print(" ")

# frekuensi dari masing-masing data di kolom host_response_time
print("frekuensi dari masing-masing nilai:")
print(listings["host_response_time"].value_counts())
```

```python
# data unik pada kolom bedrooms
data_unik_bedrooms = listings["bedrooms"].unique()
print(f"data unik pada kolom bedrooms: {data_unik_bedrooms}")

#membuat spasi
print("")

# frekuensi dari masing-masing data di kolom bedrooms
print("frekuensi dari masing-masing nilai:")
print(listings["bedrooms"].value_counts())
```

```python
# mendapatkan modus dari host_response_time
modus_host_response_time = listings.host_response_time.mode()[0]

# input missing value host_response_time dengan modus
listings.host_response_time.fillna(modus_host_response_time, inplace=True)
```

```python
# mendapatkan modus dari host_response_time
modus_bedrooms = listings.bedrooms.mode()[0]

# input missing value host_response_time dengan modus
listings.bedrooms.fillna(modus_bedrooms, inplace=True)
```

```python
# Distribusi kolom percent_host_response_rate
sns.histplot(data=listings, x="percent_host_response_rate")
```

```python
# mencari median kolom percent_host_response_rate
median_percent_host_response_rate = listings.percent_host_response_rate.median()

# menginput nan dengan median
listings.percent_host_response_rate.fillna(median_percent_host_response_rate, inplace=True)
```

```python
# Distribusi kolom percent_host_acceptance_rate
sns.histplot(data=listings, x="host_acceptance_rate")
```

```python
# mencari median kolom percent_host_response_rate
median_host_acceptance_rate = listings.host_acceptance_rate.median()

# menginput nan dengan median
listings.host_acceptance_rate.fillna(median_host_acceptance_rate, inplace=True)
```
