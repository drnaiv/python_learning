```python
# use pandas
import pandas as pd
```

```python
# import the data frame
df = pd.read_csv(r"D:\Data Sciences\Project\Audible Dataset\audible_uncleaned.csv")
df
```

```python
# data info
df.info()
```

```python
# find null values
df.isnull().sum()
```

```python
# find duplicate
df.duplicated().sum()
```

```python
# removed written by:
df['author'] = df['author'].str.replace('Writtenby:','')

# show author after removed written by:
df.author
```

```python
# give a proper author name
df['author'] = df['author'].str.replace(pat = r"(\w)([A-Z])", repl = r"\1 \2", regex = True)
# show author after given proper name
df.author
```

```python
# split author into columns
df2 = pd.concat( [df['name'], df['author'].str.split(',', expand = True).add_prefix('author'), df.loc[:,['narrator', 'time', 'releasedate', 'language', 'stars', 'price']]], axis = 1)

# author after split
df2
```

```python
# find null values
df2.loc[:,['author0','author1','author3','author3']].isnull().sum()
```

```python
# removed narrated by:
df2['narrator'] = df2['narrator'].str.replace('Narratedby:','')
```

```python
# give a proper narrator name
df2['narrator'] = df2['narrator'].str.replace(pat = r"(\w)([A-Z])", repl = r"\1 \2", regex = True)

# show author after given proper name
df2
```

```python
# find null values
df2.loc[:,'narrator'].isnull().sum()
```

```python
# copy time value
time_value = df2['time']

# replace null value with blanks
time_value = time_value.str.replace(pat = r'[0-9]', repl = '', regex = True)

# find pattern
time_value.drop_duplicates()
```

```python
# add time_hours & time_minutes
df2['time_hours'] = 0
df2['time_minutes'] = 0
```

```python
# extract hour & min from time columns to time_hours & time_minutes
df2['time_hours'] = df2.time.str.extract(pat = r'^(\d+) hr')
df2['time_minutes'] = df2.time.str.extract(pat = r'(\d+) min')
```

```python
# review extracted time

df2.loc[:,['time', 'time_hours', 'time_minutes']].sample(5)
```

```python
# replace null value with 0 from time_hours & time_minutes
df2['time_hours'] = df2['time_hours'].fillna(0)
df2['time_minutes'] = df2['time_minutes'].fillna(0)
```

```python
# change time_hours & time_minutes to int
df2 = df2.astype({'time_hours':'int64','time_minutes':'int64'})

# checking data type
df2.info()
```

```python
# uniform date format
df2.releasedate = pd.to_datetime(df2.releasedate, format = '%d-%m-%y')

# checking data type
df2.info()
```

```python
# find uniformity
df2['language'].unique()
```

```python
#create uniformity
df2.language = df2.language.str.title()
```

```python
# check uniformity
df2['language'].unique()
```

```python
# find unique rating
df2['stars'].value_counts()
```

```python
# create rating column
df2[['rating_out_of_5', 'total_rating']] = df2.stars.str.split('stars', expand = True)

# extract rating from stars column to rating_out_of_5 & total rating
df2.loc[:,['stars', 'rating_out_of_5', 'total_rating']].head(5)
```

```python
# remove out of 5 from rating_out_of_5 column
df2['rating_out_of_5'] = df2['rating_out_of_5'].str.replace(' out of 5', '')

# remove ratings from total_rating
df2['total_rating'] = df2['total_rating'].str.replace(' ratings', '')

# remove ',' from total_rating
df2['total_rating'] = df2['total_rating'].str.replace(' ratings', '')

# remove Not Rated Yet from rating_out_of_5
df2['rating_out_of_5'] = df2['rating_out_of_5'].str.replace('Not rated yet', '')
```

```python
# drop false header
df2.drop(['stars_out_of_5','total_ratings','stars'],axis=1)
```

```python
# rename column name
df2.rename(columns={'author0':'author_1','author1':'author_2', 'author2': 'author_3','author3':'author_4','releasedate':'release_date'}, inplace = True)
```

```python
# show only cleaned dataframe
cleaned_df = df2.loc[:,['name', 'author_1', 'author_2', 'author_3', 'author_4', 'narrator', 'release_date','time_hours', 'time_minutes', 'language', 'rating_out_of_5', 'total_rating', 'price']]
```

```python
# final dataframe
cleaned_df
```
